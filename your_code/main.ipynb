#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1
# your answer here

employee.groupby(['Department','Title']).agg({'Salary':np.mean}).reset_index()
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# Method 2
# your answer here

employee.pivot_table(index=['Department', 'Title'], values = ['Salary'],aggfunc = ['mean']).reset_index()
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
# your answer here

employee.pivot_table(index=['Department', 'Title','Years'], values = ['Salary'],aggfunc = ['mean']).reset_index()
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# your answer here

df = employee.groupby(['Department','Title']).agg({'Salary':np.mean,'Gender':pd.Series.nunique})
df.head()
df.columns = ['Avg-Salary', 'No.Employee']
df
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here

em_std = employee.groupby(['Department','Title']).describe()['Salary'].reset_index()
em_std
# Function describe is showing all statistics for each depertment
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
# your answer here

# STD Deviation = NaN
# STD Deviation should calculated several data, so once the data is only 1, so STD deviation is not correct.
# No Employee of HR-VP, IT-VP, IT-Associate is only 1 employee

# STD Deviation = 0
# In this case, we have 2 employee that have the same salary so the result are the same 
# as there is no range to the meadian
```

```python
# Create the function to change the STD Number 

em_std['std'].fillna('0', inplace=True)
em_std
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here

df1 = employee.groupby(['Department','Title']).agg({'Salary':'sum','Gender':pd.Series.nunique})
df1.head()
df1.columns = ['Total-Salary', 'No.Employee']
df1
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
# your answer here
df2 = employee.groupby(['Department','Title']).agg({'Years':'sum','Salary':np.mean})
df2.head()
df2.columns = ['Total-Years', 'Average-Salary']
df2
```
